```python
#Count the value counts based on labels
iris_df.value_counts('petal width (cm)')
```

```python
#Goup the value and find mean
iris_df.groupby('petal width (cm)').mean()
```

```python
#Statistical measures

# Counting no of values in each column
iris_df.count()
```

```python
# mean value column wise
iris_df.mean()
```

```python
# standard deviation column wise
iris_df.std()
```

```python
# getting min and max values
iris_df.min()
```

```python
iris_df.max()
```

```python
# get all statistical values
iris_df.describe()
```

Manipulating a df

```python
# adding a column to a dataframe
iris_df['new_column'] = iris_data.target
```

```python
iris_df.head()
```

```python
# removing a row
iris_df.drop(index=0,axis=0) # removes column with index 0
```

```python
# removing a column
# axis should be 1 if we are removing a col and 0 if we are removing a row
iris_df.drop(columns='sepal width (cm)',axis=1)
```

```python
# locating a row using index value
iris_df.iloc[2]
```

```python
# locating a particular column
print(iris_df.iloc[:,0]) # first column
print(iris_df.iloc[:,1]) # second column
print(iris_df.iloc[:,2]) # third column
print(iris_df.iloc[:,-1]) # last column
```
