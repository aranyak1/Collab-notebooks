Imbalanced class distribution:

A dataset with an unequal class distribution

```python
import numpy as np
import pandas as pd
```

```python
credit_card_data = pd.read_csv('/content/credit_data.csv')
```

```python
credit_card_data.head()
```

```python
# distribution of two classes
credit_card_data['Class'].value_counts()
```

This is highly imbalanced dataset

0 --> legit transactions

1--> fraudlent transactions

```python
# seperate the legit and fraudlent transactions
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

Undersampling

Build a sample dataset containing similar distribution of legit and fraudlent transactions

```python
legit_sample = legit.sample(n=49) # select 49 data from legit data
```

Concatenate the two dataframes

```python
new_dataset = pd.concat([legit_sample,fraud],axis = 0)
```
